#**Introducción:**

El acceso a medicamentos vitales es algo fundamental para garantizar la salud pública y el bienestar de todos. Sin embargo, la disponibilidad de estos medicamentos puede verse comprometida por distintas causas. Este análisis se enfoca en un dataset que recoge información sobre la no disponibilidad de medicamentos esenciales según diferentes diagnósticos, con el objetivo de identificar patrones y . A través de este estudio, se busca analizar los efectos y la gravedad de la falta de medicamentos utilizados para tratar diferentes enfermedades como Tumores, Diabetes, Fibrosis entre otras.

#**Conjunto de datos:**

- Ministerio de Salud y Protección Social de Colombia. (n.d.). Medicamentos vitales no disponibles. Recuperado de https://www.datos.gov.co/Salud-y-Protecci-n-Social/MEDICAMENTOS-VITALES-NO-DISPONIBLES/sdmr-tfmf/about_data

Dentro del dataset se puede presenciar una lista de diagnósticos para diferentes enfermedades con fecha y código, junto a indicación del nombre del medicamento necesario para el tratamiento junto con su concentración, forma, medida, principio activo, y cantidad solicitada por cada entidad de salud.

De modo que los datos o variables que nos interesan para nuestro analisis son las siguientes:

- **FECHA_DE_AUTORIZACIÓN**: Se detalla en esta casilla, la fecha en que se genero la autorización. (Marca de tiempo variable)

- **TIPO_DE_SOLICITUD**: Se detalla en esta casilla, si el trámite autorizado fue para un paciente específico o para varios pacientes. (Texto)

- **SOLICITANTE/IMPORTADOR**: Nombre de la empresa importador a quien se le emite la autorización de importación. (Texto)

- **IUM**: Identificador único de medicamento, acorde a la Resolución 3166 de 2015. (Texto)

- **PRINCIPIO_ACTIVO1**: Princinpio activo del medicamento autorizado. (Texto)

- **CONCENTRACIÓN_DELMEDICAMENTO1**: Relación entre la cantidad de soluto y la cantidad de disolución o de disolvente. Puede ser expresada en peso o volumén. (Texto)

- **UNIDAD_MEDIDA1**: Cantidad estandarizada expresada en el sistema internacional de medida. (Texto)

- **PRINCIPIO_ACTIVO2**: En casos que aplica, es decir, combinaciones de principio activo en un mismo medicamento. (Texto)

- **CONCENTRACIÓN_DEL_MEDICAMENTO2**: Relación entre la cantidad de soluto y la cantidad de disolución o de disolvente. Puede ser expresada en peso o volumén. (Texto)

- **UNIDAD_MEDIDA2**: Cantidad estandarizada expresada en el sistema internacional de medida. (Texto)

- **FORMA_FARMACÉUTICA**: Forma final del medicamento, directamente relacionado con la vía de administración, fabricación y dosificación. (Texto)

- **NOMBRE_COMERCIAL_**: Nombre con el que se comercializa el producto. (Texto)

- **CANTIDAD_SOLICITADA**: Número de unidades aprobadas. (Número)

- **PRESENTACIÓN_COMERCIAL**: Número de unidades de contenido en un mismo empaque. (Texto)

- **DIAGNOSTICO_CIE-1NO REPORTA**: Enfermedad o patología para la que indicando el medicamento autorizado. (Texto)
- **CÓDIGO_DIAGNOSTICO_CIE-10**: Código del diagnostico en CIE10 c_digo_diagnostico_cie_10 (Texto)

##Mal Tuning Cross validation (pero no muestra el mejor y toca uno por uno)

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer

def show_curve_cv(est, X, y):
    means, stds = [], []
    nfolds_range = range(2, 10)

    for nfolds in nfolds_range:
        s = cross_val_score(est, X, y, cv=KFold(nfolds, shuffle=True), scoring=make_scorer(accuracy_score))
        means.append(np.mean(s))
        stds.append(np.std(s))

    means = np.r_[means]
    stds = np.r_[stds]

    plt.plot(nfolds_range, means, label="mean accuracy", color="black")
    plt.fill_between(nfolds_range, means - stds, means + stds, color="blue", alpha=0.5, label="std")
    plt.xlabel("n folds")
    plt.ylabel("accuracy")
    plt.legend()
    plt.title("Learning Curve (Accuracy)")
    plt.show()
```

```python
show_curve_cv(DecisionTreeClassifier(),X,y)
```

##Mal Tuning traintestsplit (no saca el mejor y solo va 1 por 1)

```python
def show_curve_ttsplit(est, X, y):
    means, stds = [], []
    test_sizes = np.linspace(0.1, 0.5, 10)  # entre el 10 y 50%, dividido en 10 valores.

    for test_size in test_sizes:
        scores = []

        for _ in range(5): #5 pruebas
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=True)
            est.fit(X_train, y_train)
            y_pred = est.predict(X_test)
            scores.append(accuracy_score(y_test, y_pred))

        means.append(np.mean(scores))
        stds.append(np.std(scores))

    means = np.r_[means]
    stds = np.r_[stds]

    plt.plot(test_sizes, means, label="mean accuracy", color="black")
    plt.fill_between(test_sizes, means - stds, means + stds, color="blue", alpha=0.5, label="std")
    plt.xlabel("test size")
    plt.ylabel("accuracy")
    plt.legend()
    plt.title("Learning Curve (Accuracy vs. Test Size)")
    plt.show()
```

```python
show_curve_ttsplit(DecisionTreeClassifier(),X,y)
```

##Tuning de parametros en un estimador
Busqueda de que parametros dan el MEJOR acuracy
- DecisionTree (max_depth, criterion)
- RandomForest (n_estimators, criterion)
- SVC/R (kernel, gamma)


```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def show_decision_tree_tts(X, y):
    max_depth_values = range(10, 101, 10)  # max_depth de 10 a 100
    criteria = ['gini', 'entropy']

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

    results = {criterion: [] for criterion in criteria}  # diccionario para almacenar los scores
    best_params = {"criterion": None, "max_depth": None, "accuracy": 0}  # inicializar un diccionario para guardar los mejores accuracy

    for criterion in criteria:
        for max_depth in max_depth_values:
            # entrenar el modelo con cada cosa de la lista criteria
            est = DecisionTreeClassifier(max_depth=max_depth, criterion=criterion)
            est.fit(x_train, y_train)

            y_pred = est.predict(x_test)
            score = accuracy_score(y_test, y_pred)
            results[criterion].append(score)

            if score > best_params["accuracy"]: #calculando los mejores scores
                best_params = {"criterion": criterion, "max_depth": max_depth, "accuracy": score}

    print(f"Mejores parámetros encontrados:")
    print(f"  Criterion: {best_params['criterion']}")
    print(f"  Max Depth: {best_params['max_depth']}")
    print(f"  Accuracy: {best_params['accuracy']:.4f}")

    plt.figure(figsize=(10, 6))
    for criterion, scores in results.items():
        plt.plot(max_depth_values, scores, label=f"criterion={criterion}")

    plt.xlabel("max_depth")
    plt.ylabel("Accuracy")
    plt.title("Decision Tree Accuracy vs max_depth")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
show_decision_tree_tts(X,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import LabelEncoder

X = df.drop(["NOMBRE_COMERCIAL_", "IUM"], axis=1)
y = df['NOMBRE_COMERCIAL_']

X = X.dropna()
y = y.loc[X.index]

X = pd.get_dummies(X, drop_first=True)

if y.dtype == 'object' or y.dtype.name == 'category':
    le = LabelEncoder()
    y = le.fit_transform(y)

X = X[:500]
y = y[:500]

def tuneado_cv(estimators, X, y, fold_range=range(5, 8)):
    plt.figure(figsize=(8, 5))
    for name, estimator_class in estimators.items():
        mean_accuracies = [
            np.mean(cross_val_score(estimator_class(), X, y, cv=KFold(n_splits=f, shuffle=True, random_state=42), n_jobs=-1))
            for f in fold_range
        ]
        plt.plot(fold_range, mean_accuracies, marker='o', label=name)
        best_folds = fold_range[np.argmax(mean_accuracies)]
        print(f"Mejor número de folds para {name}: {best_folds} (Precisión: {max(mean_accuracies):.4f})")
    plt.xlabel("Número de Folds")
    plt.ylabel("Precisión Media")
    plt.title("Precisión vs Número de Folds")
    plt.legend()
    plt.grid(True)
    plt.show()

estimators = {
    "Decision Tree": DecisionTreeClassifier,
    "Random Forest": lambda: RandomForestClassifier(n_estimators=10, max_depth=5),
    "SVC": lambda: SVC()
}

tuneado_cv(estimators, X, y)
```

```python
medicamentos = df['NOMBRE_COMERCIAL_'].unique()
demanda = df.groupby('NOMBRE_COMERCIAL_')['CANTIDAD_SOLICITADA'].sum().tolist()

NUM_GENERACIONES = 50
TAM_POBLACION = 20
PROB_MUTACION = 0.1
PROB_CRUCE = 0.8


def calcular_fitness(individuo):
    """Calcula la calidad de una solución (individuo)."""
    cobertura = sum(min(individuo[i], demanda[i]) for i in range(min(len(individuo), len(demanda))))
    return cobertura

def generar_poblacion(tam_poblacion, num_medicamentos):
    return [np.random.randint(0, max(demanda) + 1, size=num_medicamentos) for _ in range(tam_poblacion)]

def seleccionar_padres(poblacion, fitness):
    indices = np.random.choice(len(poblacion), size=2, replace=False)
    if fitness[indices[0]] > fitness[indices[1]]:
        return poblacion[indices[0]]
    else:
        return poblacion[indices[1]]

def cruzar(padre1, padre2):
    if np.random.rand() < PROB_CRUCE:
        punto_cruce = np.random.randint(1, len(padre1))
        hijo1 = np.concatenate([padre1[:punto_cruce], padre2[punto_cruce:]])
        hijo2 = np.concatenate([padre2[:punto_cruce], padre1[punto_cruce:]])
        return hijo1, hijo2
    else:
        return padre1.copy(), padre2.copy()

def mutar(individuo):
    for i in range(len(individuo)):
        if np.random.rand() < PROB_MUTACION:
            individuo[i] = np.random.randint(0, max(demanda) + 1)
    return individuo

def algoritmo_genetico():
    poblacion = generar_poblacion(TAM_POBLACION, len(medicamentos))
    mejor_fitness_por_generacion = []

    for generacion in range(NUM_GENERACIONES):
        fitness = np.array([calcular_fitness(ind) for ind in poblacion])
        mejor_fitness_por_generacion.append(fitness.max())

        nueva_poblacion = []
        for _ in range(TAM_POBLACION // 2):
            padre1 = seleccionar_padres(poblacion, fitness)
            padre2 = seleccionar_padres(poblacion, fitness)
            hijo1, hijo2 = cruzar(padre1, padre2)

            hijo1 = hijo1[:len(demanda)]
            hijo2 = hijo2[:len(demanda)]

            nueva_poblacion.append(mutar(hijo1))
            nueva_poblacion.append(mutar(hijo2))

        poblacion = nueva_poblacion

    fitness = np.array([calcular_fitness(ind) for ind in poblacion])
    mejor_individuo = poblacion[fitness.argmax()]

    plt.plot(mejor_fitness_por_generacion)
    plt.title("Progreso del fitness por generación")
    plt.xlabel("Generación")
    plt.ylabel("Fitness")
    plt.show()

    return mejor_individuo, fitness.max()

mejor_solucion, mejor_fitness = algoritmo_genetico()
```
